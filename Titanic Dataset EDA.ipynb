```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Titanic Dataset Python.csv')
```

```python
print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nValue Counts (Categorical Columns):")
print("Survived:\n", df['Survived'].value_counts())
print("Pclass:\n", df['Pclass'].value_counts())
print("Sex:\n", df['Sex'].value_counts())
```

```python
sns.pairplot(df, hue='Survived')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.hist(figsize=(12,10))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Age Distribution by Passenger Class")
plt.show()

plt.figure(figsize=(8,6))
sns.boxplot(x='Sex', y='Fare', data=df)
plt.title("Fare Distribution by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare")
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title("Class vs Fare")
plt.show()
```
